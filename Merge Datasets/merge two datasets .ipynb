##### df1 has 10 rows and df2 has 11 rows

```python
merge_df = pd.merge(df1, df2, on = 'use_id')
merge_df
```

```python
merge_df.shape
```

#### Why one row's data missing??

By default, pandas use INNER JOIN for this criteria. Please see the below picture you will figure out the concept.
<img src="https://www.ionos.com/digitalguide/fileadmin/DigitalGuide/Screenshots_2018/Outer-Join.jpg">

```python
merge_df = pd.merge(df1, df2, on = 'use_id', how='outer') #By default, how='inner'
merge_df.tail(5)
```

```python
merge_df.shape
```

#### Now it's working :)

##### you can add pirority in how=' right'/how='left' ; df1/left>df2/right

## How='left'

```python
merge_df = pd.merge(df1, df2, on = 'use_id', how='left')
merge_df.tail(5)
```

## How='right'

```python
merge_df = pd.merge(df1, df2, on = 'use_id', how='right')
merge_df.tail(5)
```

#### Add indicator for merge information

```python
merge_df = pd.merge(df1, df2, on = 'use_id', how='right', indicator = True)
merge_df.tail(5)
```

##### indicator=True is create new columns which name 'is _merge'
