#### Importing Libraries

```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse=False, drop='first')
df = pd.read_excel('/Dataset/district-wise-confirmed.2020.xlsx')
```

##### Fit desire columns for One Hot Encoding

```python
df_ohe = pd.DataFrame(ohe.fit_transform(df[['Division']]))
df_ohe
```

```python
df = df.join(df_ohe)
df 
```
